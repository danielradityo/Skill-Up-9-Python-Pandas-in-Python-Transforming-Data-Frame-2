```python
superstore = pd.read_csv('Superstore-1000.csv').iloc[:, 1:]
superstore.head(2)
```

```python
by_cat = superstore.groupby(['Category', 'Sub-Category']).sum()[['Sales', 'Profit']].reset_index()
```

```python
by_cat.sort_values('Profit', ascending=False)
```

```python
by_city = superstore.groupby(['City']).agg({'Sales' : 'sum', 'Profit' : 'mean'}).reset_index()
by_city = by_city.rename(columns = {'Sales' : 'Total Sales', 'Profit' : 'Average Profit'})
by_city
```

```python
by_city = superstore.groupby(['City']).agg({'Sales' : 'sum', 'Profit' : ['sum', 'mean']}).reset_index()
by_city = by_city.droplevel(level=0, axis=1)
by_city.columns = ['City', 'Total Sales', 'Average Profit', 'Total Profil']
by_city
```

## Menggunakan Lambda

```python
def segment_sales(x):
    if x['Sales'] < 200:
        segment = 'Rendah'
    elif 200 <= x['Sales'] < 1000:
        segment = 'Sedang'
    else:
        segment = 'Tinggi'
    return segment

data['Sales Segment'] = data.apply(lambda x: segment_sales(x), axis=1)
```
